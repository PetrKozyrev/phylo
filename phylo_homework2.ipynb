```python
#1. читает дерево https://www.jasondavies.com/tree-of-life/life.txt
#2. рисует это дерево псевдографикой (draw_ascii);
from Bio import Phylo
from matplotlib import pylab


tree1 = Phylo.read("/home/petr/bioinformatics/ib/R/phylo/life.txt", "newick")
Phylo.draw_ascii(tree1)
```

```python
# 3. рисует дерево с помощью draw;
Phylo.draw(tree1)
```

```python
# 4. сохраняет изображение дерева в растровый формат (png) и векторный (svg/pdf)

(можно с помощью pylab.savefig, например) (картинки опять прислать);
Phylo.draw(tree1)
pylab.savefig("tree1.png", dpi=200)
pylab.savefig("tree1.svg")
```

```python
#7. читает дерево
from ete3 import Tree

tree2 = Tree("/home/petr/bioinformatics/ib/R/phylo/life.txt", format=1)
```

```python
#8. вырезает (функция prune) из дерева случайный набор из 42 (можно другое число)
# листьев (можно не случайный набор)
import random
random.seed(100)
ss = random.sample(tree2.get_leaf_names(), 42)
tree2.prune(ss)
```

```python
#9. рисует любое (на выбор) из этих деревьев (можно оба)
tree2.show()
```

```python
#10. рисует любое (на выбор) из этих деревьев с дополнительной обработкой символов
# на узлах (в смысле нужно поменять дефолтные синие точки на что-нибудь)
from ete3 import NodeStyle

nstyle = NodeStyle()
nstyle["shape"] = "sphere"
nstyle["size"] = 10
nstyle["fgcolor"] = "darkred"

for n in tree2.traverse():
   n.set_style(nstyle)
tree2.show()
```
